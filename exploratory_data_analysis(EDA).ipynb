#Paso 1

para empezar con esta explicacion primero descargaremos la data

```python
#primero importamos la libreria necesaria
import pandas as pd

#leemos la data con pd.read_csv
df= pd.read_csv("datos_paises.csv")
```

#paso 2
el primer paso en el analisis exploratorio es revisar el contenido

```python
#ver las primeras 5 filas del dataframe, esto se hace para poder ver una vista previa
df.head()

#tambien puedes usar sample para ver valores random
df.sample(5)
```

```python
#ver los valores de la media, desviacion media, min, 1/4 2/4 3/4 y max de los valores numericos
df.describe()
```

```python
#ver el contenido de una columna con unique, en este caso los paises
df['pais'].unique()
#en este caso el continente
df['continente'].unique()
#contar los valores unicos
#df['pais'].nunique()
```

```python
#contar valores de una columna
df['pais'].value_counts()
df['km2'].value_counts()
```

#paso 3
ver la distribucion de los datos con graficos

```python
#grafico de pastel
import matplotlib.pyplot as plt
plt.pie(df['continente'].value_counts(), labels=['Asia', 'America', 'Africa', 'Europa', 'Oceania'], autopct= "%0.2f")
plt.show()
```

```python
#ver el histograma
import seaborn as sns
sns.histplot(df[df['continente']== 'Africa']['km2'])
sns.histplot(df[df['continente']== 'Europa']['km2'])
sns.histplot(df[df['continente']== 'Oceania']['km2'])
sns.histplot(df[df['continente']== 'Asia']['km2'])
sns.histplot(df[df['continente']== 'America']['km2'])
plt.show()
```

```python
#ver distintos graficos de dispersion de los datos
sns.pairplot(df, hue='continente')
```

```python
sns.heatmap(df.corr(), annot= True)
plt.show()
```
